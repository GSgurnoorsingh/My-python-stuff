```python
import matplotlib.pyplot as plt

scores = [56, 78, 45, 80, 92, 67, 85, 66, 73, 89]
plt.hist(scores, bins=5, color='skyblue')
plt.title("Histogram of Test Scores")
plt.xlabel("Score Range")
plt.ylabel("Frequency")
plt.show()
```

```python
hours = [1, 2, 3, 4, 5, 6, 7]
scores = [50, 55, 60, 65, 70, 75, 80]
plt.scatter(hours, scores, color='green')
plt.title("Study Hours vs Exam Scores")
plt.xlabel("Hours Studied")
plt.ylabel("Score")
plt.show()
```

```python
products=['Apples','Banana','Cherries']
sales=[120,90,150]
plt.bar(products,sales,color='orange')
plt.title("Sales per product")
plt.xlabel("Product")
plt.ylabel("Units Sold")
plt.show()
```
